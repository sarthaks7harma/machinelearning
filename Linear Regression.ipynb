```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
noise=np.random.randn(100)
noise.shape
```

```python
x=np.linspace(0,10,100)
x.shape
```

```python
y=2*x+noise
x=(x-x.mean())/x.std()
plt.scatter(x,y)
plt.show()
```

```python
def hypothesis(x,theta):
    return theta[0]+theta[1]*x


def error(x,y,theta):
    err=0
    m=x.shape[0]
    for i in range(m):
        hx=hypothesis(x[i],theta)
        err=err+(hx-y[i])**2
    return err
def grad(x,y,theta):
    g=np.array([0,0])
    for i in range (x.shape[0]):
        hx=hypothesis(x[i],theta)
        g[0]+=hx-y[i]
        g[1]+=(hx-y[i])*x[i]
    return g
    
def gradientdescent(x,y,learningrate=0.001):
    theta=np.array([0.1,0.2])
    errorlist=[]
    for i in range(x.shape[0]):
        g=grad(x,y,theta)
        err=error(x,y,theta)
        errorlist.append(err)
        theta[0]=theta[0]-learningrate*g[0]
        theta[1]=theta[1]-learningrate*g[1]
    return theta,errorlist    
```

```python
theta,err=gradientdescent(x,y)
plt.plot(err)
print(theta)
```

```python
plt.scatter(x,y)
xtest=np.linspace(-2,2,10)
plt.plot(xtest,hypothesis(xtest,theta),c="orange",label="test")
plt.legend()
plt.title("linear regression")
plt.show()
```
